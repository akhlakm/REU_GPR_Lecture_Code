```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.rc("font", size=16)
```

```python
x = np.random.normal(scale=1.0, size=10000)
```

```python
plt.hist(x, bins=100)
plt.show()
```

```python
y = np.random.normal(scale=2.0, size=10000)
z = np.random.normal(scale=3.0, size=10000)
plt.hist(z, bins=100, alpha=0.6, color='r')
plt.hist(y, bins=100, alpha=0.4, color='b')
plt.hist(x, bins=100, alpha=0.6, color='k')
plt.show()
```

```python
plt.plot(z, 'r.')
plt.plot(y, 'b.')
plt.plot(x, 'k.')
plt.show()
```

```python
plt.figure(figsize=(8,8))
plt.plot(x, y/2+2*x, 'k.')
plt.xlabel("x")
plt.ylabel("z")
plt.xlim(-12, 12)
plt.ylim(-12, 12)
plt.grid()
plt.show()
```

```python
## PV = n R T
n = 10 # mol
R = 8.316 # J/mol-K

T = 273 + np.random.normal(loc=27, scale=2, size=500) # K
V = np.random.normal(loc=100, scale=1, size=500) # m^3
```

```python
plt.plot(T, 'k.')
plt.show()
```

```python
P = n * R * T / V
```

```python
plt.plot(V, P, 'k.')
plt.show()
```
